```python
import pandas as pd
import numpy as snp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax = plt.subplots()
ax.set_xlabel('жилищный фонд , млн.кв.м   (2005г)')
ax.set_ylabel('Количество регионов')
plt.hist(data["jf2005"], bins=[0,50,100,150,200,250,300,350])
plt.show()
```

```python
fig, ax = plt.subplots()
ax.set_xlabel('жилищный фонд , млн.кв.м   (2020г)')
ax.set_ylabel('Количество регионов')
plt.hist(data["jf2020"], bins=[0,50,100,150,200,250,300,350])
plt.show()
```

```python
index ='распределение 2005'
data[index] = pd.cut(data['jf2005'], [0,50,100,150,200,250,300,350], labels=[1,2,3,4,5,6,7])
regions = []
clusters = [1,2,3,4,5,6,7]
for cluster in clusters:
  regions = data[(data[index] == cluster)].Регион
  cluster = cluster *50
  print(f'В интервал от {cluster - 50} до {cluster} входят регионы:\n',regions)
```
